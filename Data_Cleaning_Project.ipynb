```python
# Convert 'Not Known' values to NaN
df["Phone"] = df["Phone"].replace("Not Known", np.nan)
# Convert the column to numeric, coercing errors to NaN
df["Phone"] = pd.to_numeric(df["Phone"], errors='coerce')
# Calculate the median of the numeric values
x = df["Phone"].median()
# Fill the NaN values with the calculated median
df["Phone"].fillna(x, inplace=True)
```
